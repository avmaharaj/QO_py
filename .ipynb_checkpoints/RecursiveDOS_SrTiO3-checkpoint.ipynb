# Quantum Oscillations - python implementation

** A six band model of SrTiO3, with weak spin orbit coupling**

```python
#Import the external modules, and set up the notebook environment
from solver import RecursiveSolver
from hamiltonian import T2g_SpinOrbit_Model
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

# for auto-reloading external modules
# see http://stackoverflow.com/questions/1907993/autoreload-of-modules-in-ipython
%load_ext autoreload
%autoreload 2
```

```python
#Initialize the model
mymodel = T2g_SpinOrbit_Model( spin_orbit = 0.0, theta = 0.0, phi = 1.1, mu = -4.35, delta = 0.001, t2 =0.2)

#Initialize the solver 
solver = RecursiveSolver(mymodel, system_size = 2048, min_invfield = 10.0, max_invfield = 10000.0,num_steps = 5000)

#Run the code!
out = solver.run()
```

```python
#Plot the results - oscillations of the density of states vs. 1/B
plt.plot(out[:,0].real,out[:,1].real,'-o')
```

```python
#Perform a Fourier transform of the data
b = np.fft.fft(out[:,1].real)
n = out[:,0].size
invbstep = out[1,0]-out[0,0]
freq = np.fft.fftfreq(n, d=invbstep)
plt.plot(freq,np.abs(b),'o-')
plt.xlim((0.001,0.1))
plt.ylim((0,1000))
```
